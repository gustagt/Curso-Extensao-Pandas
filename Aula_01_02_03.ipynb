```python
# Para instalar, use esse comando!
# !pip install pandas

#Importando o Pandas
import pandas as pd

combustiveis_df = pd.read_excel('ca-2021-02.xlsx')
```

```python
# Usa o print para ver o dataframe!
display(combustiveis_df)
```

```python
display(combustiveis_df.head(30))
```

```python
# Mostra as linhas e colunas
display(combustiveis_df.shape)
```

```python
# Mostra os principais calculos
display(combustiveis_df.describe())
```

```python
#Quais são as colunas e cada tipo de dados ela recebe
print(combustiveis_df.info())
```

```python
# filtrando coluna
display(combustiveis_df['Revenda'][0:10])
```

```python
# filtrando colunas
novo_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]

display(novo_df)
```

```python
# função para localizar
display(novo_df.loc[3])
```

```python
# ordenações
display(novo_df.loc[9:19])
display(novo_df.loc[9::19])
display(novo_df.loc[9:3:19])
display(novo_df.loc[9::])
display(novo_df.loc[::9])
display(novo_df.loc[::-1])
```

```python

# filtro pelo valor da coluna
gas_df = novo_df.loc[novo_df['Produto'] == 'GASOLINA']

display(gas_df)
```

```python
#o maior exemplo da coluna
display(gas_df['Valor de Venda'].max())
```

```python
#o max relaciona a coluna que tiver o tipo da coluna numerica
print(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#filtragem com loc
etanol_indaiatuba_df = novo_df.loc[(novo_df['Produto'] == 'ETANOL') & (novo_df['Municipio'] == 'CONTAGEM')]
etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True)
display(etanol_indaiatuba_df)
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | 
                             (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL Brasil todo?
media_por_combustivel_df = novo_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
#Quero adicionar uma coluna de valor booleano no combustiveis_df
# chamada "Ativo" que, por padrão, vai ser True para 
# TODAS as linhas
combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```
